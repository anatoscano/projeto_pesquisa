### Funções e Numeros de operações T(n) 

```python
def ler_numeros_do_arquivo(nome_ficheiro):
    with open(nome_ficheiro, 'r') as ficheiro:
        numeros = [int(line.strip()) for line in ficheiro]
    return numeros
```

### 1. Pesquisa Sequencial

```python
def pesquisa_sequencial(lista,n):
    operacoes = 0
    for item in lista:
        operacoes += 1
        if item == n:
            return operacoes
    return operacoes
```

### 2. Pesquisa Sequencial Recursiva

```python
def pesquisa_sequencial_recursiva(lista, n, operacoes = 0):
    if not lista:
        return operacoes
    operacoes += 1
    if lista[0] == n:
        return operacoes
    operacoes = pesquisa_sequencial_recursiva(lista[1:], n, operacoes)
    return operacoes
```

### 3. Pesquisa Binaria

```python
def pesquisa_binaria(lista,n):
    operacoes = 0
    primeiro = 0
    ultimo = len(lista) - 1
    while primeiro <= ultimo:
        operacoes += 1
        meio = (primeiro + ultimo)//2
        if lista[meio] == n:
            return operacoes
        elif n < lista[meio]:
            ultimo = meio - 1
        else:
            primeiro = meio + 1
    return operacoes
```

### 4. Pesquisa Binaria Recursiva

```python
def pesquisa_binaria_recursiva(lista, n, operacoes = 0):
    if len(lista) == 0:
        return operacoes
    operacoes += 1
    meio = len(lista)//2
    if lista[meio] == n:
        return operacoes
    elif n < lista[meio]:
        operacoes = pesquisa_binaria_recursiva(lista[:meio], n, operacoes)
    else:
        operacoes = pesquisa_binaria_recursiva(lista[meio+1:], n, operacoes)
    return operacoes
```

### 4.1. Complexidade Temporal 

```python
def gerar_sequencia_ordenada(tamanho):
    return [random.randint(1, 1000000) for _ in range(tamanho)]

tamanhos = [100, 1000, 5000, 10000, 50000]
```

#### 4.1.1. a) ordenados

```python
tempos_binaria_recursiva_ordenada = []
operacoes_binaria_recursiva_ordenada = []

for tamanho in tamanhos:
    sequencia = gerar_sequencia_ordenada(tamanho)
    alvo = random.choice(sequencia)

    sequencia.sort()
    tempo_inicial = time.time()
    operacoes = pesquisa_binaria_recursiva(sequencia, alvo)
    tempo_final = time.time()
    tempo_decorrido = tempo_final - tempo_inicial

    tempos_binaria_recursiva_ordenada.append(tempo_decorrido)
    operacoes_binaria_recursiva_ordenada.append(operacoes)

plt.figure(figsize=(6, 6))
plt.plot(tamanhos, tempos_binaria_recursiva_ordenada, label='Pesquisa Binária Recursiva (Sequências Ordenadas)')
plt.xlabel('Tamanho da Sequência')
plt.ylabel('Tempo (segundos)')
plt.title('Complexidade Temporal - Pesquisa Binária Recursiva em Sequências Ordenadas')
plt.legend()

plt.show()
```

#### 4.1.2. b) não ordenados

```python
tempos_binaria_recursiva_ordenada = []
operacoes_binaria_recursiva_ordenada = []

for tamanho in tamanhos:
    sequencia = gerar_sequencia_ordenada(tamanho)
    alvo = random.choice(sequencia)

    tempo_inicial = time.time()
    operacoes = pesquisa_binaria_recursiva(sequencia, alvo)
    tempo_final = time.time()
    tempo_decorrido = tempo_final - tempo_inicial

    tempos_binaria_recursiva_ordenada.append(tempo_decorrido)
    operacoes_binaria_recursiva_ordenada.append(operacoes)

plt.figure(figsize=(6, 6))
plt.plot(tamanhos, tempos_binaria_recursiva_ordenada, label='Pesquisa Binária Recursiva (Sequência Não Ordenada)')
plt.xlabel('Tamanho da Sequência')
plt.ylabel('Tempo (segundos)')
plt.title('Complexidade Temporal - Pesquisa Binária Recursiva (Sequência Não Ordenada)')
plt.legend()

plt.show()
```
